```python
cars_address = '/Users/labattula/Documents/Lakshman/Future-ML/MyDS/Ex_Files_Python_Data_Science_EssT/Exercise Files/Ch01/01_05/mtcars.csv'
cars = pd.read_csv(cars_address)

cars.columns = ['car_names','mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb']

cars.head()
```

```python
mpg = cars.mpg
mpg
```

```python
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

mpg.plot()
ax.set_xticks(range(32))

ax.set_xticklabels(cars.car_names,rotation=60,fontsize='small')
ax.set_title('Cars with miles per gallon chart')

ax.set_xlabel('Car Names')
ax.set_ylabel('Miles per gallon')
```
